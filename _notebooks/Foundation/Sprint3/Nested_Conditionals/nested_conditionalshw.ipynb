Objective:
Write a JavaScript program that randomly assigns a student grade level (from 0 to 12) and outputs the school level the student would be in (kindergarten, elementary, middle, or high school). The program also identifies if the student is in their last year at their current school level.

Instructions:

Open a JavaScript file (e.g., schoolLevel.js).

Generate a Random Grade:

Use Math.random() to generate a random number between 0 to 1 and multiply it by 12, which will represent the student’s grade.
Round the number to the nearest whole number using Math.round().
Store this number in a variable called grade.
Print the grade to the console with a message such as, "You are in ___ grade."
Determine the School Level:

Use if statements to check the value of grade and print which school level the student is in:
Kindergarten for grade 0
Elementary School for grades 1–5
Middle School for grades 6–8
High School for grades 9–12
For each school level, add an additional check to see if the student is in their final year. If all is done correctly, it will print a message like this, “You will graduate this year from _____.”



- Booleans are a data type that can only have two different values True or Flase. 
- Random generates random values with unpredictable results, usually within a specified range such as 0-1

```python
%%html

<html lang="en">
    <head>
        <meta charset="UTF-8">
        <meta name="viewport" content="width=device-width, initial-scale=1.0">
        <title>Student Grade Communication</title>
        <style>
            #dialogueBox {
                margin-top: 20px;
                padding: 20px;
                border: 1px solid black;
                width: 300px;
                background-color: #f0f0f0;
                color: black; /* Text color set to black */
            }

            button {
                margin-top: 20px;
                padding: 10px 20px;
                cursor: pointer;
            }

            #responseBox {
                margin-top: 20px;
            }
        </style>
    </head>
    <body>
        <h1>Student Grade Communication</h1>
        <button id="startButton">Start Grade Conversation</button>

        <div id="dialogueBox" style="display: none;"></div>

        <div id="responseBox" style="display: none;">
            <label for="gradeChoice">Enter your grade (0-12): </label>
            <input type="number" id="gradeChoice" min="0" max="12" />
            <button id="submitGrade">Submit Grade</button>
        </div>

        <script>
            // Function to start the conversation about grades
            function startConversation() {
                // Display the initial prompt asking for the grade
                document.getElementById("dialogueBox").textContent = 
                    "Please enter your grade (0-12) to determine your school level.";
                document.getElementById("dialogueBox").style.display = "block";

                // Show the input field and submit button for the player's grade
                document.getElementById("responseBox").style.display = "block";
            }

            // Function to process the player's grade and display appropriate school level
            function processGrade() {
                // Get the player's grade from the input field
                let grade = document.getElementById("gradeChoice").value;
                let message = "";

                // Determine the school level based on the grade
                switch (parseInt(grade)) {
                    case 0:
                        message = "You are in Kindergarten. You will graduate this year from Kindergarten.";
                        break;
                    case 1:
                    case 2:
                    case 3:
                    case 4:
                    case 5:
                        message = "You are in Elementary School.";
                        if (grade == 5) {
                            message += " You will graduate this year from Elementary School.";
                        }
                        break;
                    case 6:
                    case 7:
                    case 8:
                        message = "You are in Middle School.";
                        if (grade == 8) {
                            message += " You will graduate this year from Middle School.";
                        }
                        break;
                    case 9:
                    case 10:
                    case 11:
                    case 12:
                        message = "You are in High School.";
                        if (grade == 12) {
                            message += " You will graduate this year from High School.";
                        }
                        break;
                    default:
                        message = "Sorry, I didn't understand that. Please enter a grade between 0 and 12.";
                }

                // Show the response in the dialogue box
                document.getElementById("dialogueBox").textContent = message;

                // Hide the input field and submit button after the response is processed
                document.getElementById("responseBox").style.display = "none";
            }

            // Event listener for the "Start Grade Conversation" button
            document.getElementById("startButton").addEventListener("click", function() {
                startConversation();
            });

            // Event listener for the "Submit Grade" button
            document.getElementById("submitGrade").addEventListener("click", function() {
                processGrade();
            });
        </script>

    </body>
</html>
```
